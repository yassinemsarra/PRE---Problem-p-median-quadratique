```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Changer le nom du fichier benchmark 
df = pd.read_csv("../results/benchmark/benchmark_20260604_130403.csv")
df.replace(-1.0, np.nan, inplace=True)

methods  = df['method'].unique().tolist()
n_methods = len(methods)
df['instance'] = df.groupby(['n_clients','n_sites','p']).cumcount() // n_methods + 1
df['size'] = df['n_sites'].astype(str) + ' sites\n(' + df['n_clients'].astype(str) + ' clients)'
print('Chargé :', len(df), 'lignes')
```

```python
rows = []
for (nc, ns, p, inst), grp in df.groupby(['n_clients','n_sites','p','instance'], sort=False):
    row = {'Instance': f'nc={nc}, ns={ns}, p={p} | #{inst}', 'Métrique': ''}
    for method in methods:
        row[method] = ''
    rows.append(row)

    for metric, label, fmt_fn in [
        ('val_relaxation', 'Relax',  lambda v: f'{v:.1f}'),
        ('objective',      'Obj',    lambda v: f'{v:.1f}'),
        ('gap',            'Gap',    lambda v: f'{v:.2f}%'),
        ('nodes',          'Noeuds', lambda v: f'{int(v)}'),
        ('t_solve',        'Temps',  lambda v: f'{v:.4f}s'),
    ]:
        row = {'Instance': '', 'Métrique': label}
        for method in methods:
            sub = grp[grp['method'] == method]
            if sub.empty or pd.isna(sub.iloc[0][metric]):
                row[method] = '—'
            else:
                row[method] = fmt_fn(sub.iloc[0][metric])
        rows.append(row)

table = pd.DataFrame(rows)
table = table.set_index(['Instance', 'Métrique'])

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
table

```
